```python
data = pd.read_csv("chaska_cafe_sales_data.csv")
data.head()
```

```python
## Like this we can change th name of the column
data = data.rename(columns={"Sales" : "Item_Sales"})
data.head()
```

```python
## Change Datatype
data['Change_Type'] = data['Item_Sales'].fillna(data['Item_Sales'].mean()).astype(int)
data.head()
```

```python
## With lambda function
data['New Column'] = data['Item_Sales'].apply(lambda x : x * 2)
data.head()
```

## Aggregating and Grouping Data

Data aggregation and grouping allow you to compute statistics based on categorical variables. Use the `groupby()` method to group data and apply aggregation functions like `mean`, `sum`, or `count`.

```python
data.head()
```

```python
data_grouping = data.groupby('Category')['Item_Sales'].mean()
print(data_grouping)
```

### Grouping by two columns

```python
groupByTwo = data.groupby(['Category','Item'])['Item_Sales'].sum()
print(groupByTwo)
```

```python
groupByTwo = data.groupby(['Category','Item'])['Item_Sales'].mean()
print(groupByTwo)
```

```python
group_agg = data.groupby('Item')['Item_Sales'].agg(['sum','mean','count'])
print(group_agg)
```

```python
group_agg = data.groupby(['Category','Item'])['Item_Sales'].agg(['sum','mean','count'])
print(group_agg)
```

## Merging and Joining DataFrames

Merging and joining DataFrames are essential for combining datasets. You can use the `pd.merge()` function to merge two DataFrames on a specific key column. Different types of joins are available: inner, outer, left, and right joins.

```python
data1 = pd.DataFrame({'Key':['A','B','C'], 'Value':[1,2,3]})
data2 = pd.DataFrame({'Key':['A','B','D'], 'Value':[4,5,6]})
mergedInner = pd.merge(data1,data2, on='Key', how='inner') ## "on" is used for that by which thing want to megred and "how" is used for that in direction you wanna match "inner" is used for left, right
mergedInner
```

```python
mergedInner = pd.merge(data1,data2, on='Key', how='outer')
mergedInner
```

```python
mergedInner = pd.merge(data1,data2, on='Key', how='left')
mergedInner
```
