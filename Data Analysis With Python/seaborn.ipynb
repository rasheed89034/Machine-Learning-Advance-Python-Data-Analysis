## Data Visualization With Seaborn
##### Seaborn, which is used for data visualization in Python. Seaborn is a Python visualization library based on matplotlib. It provides a higher-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics to advanced features of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
```

```python
## With scatter plot
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
## now with line plot
sns.lineplot(x="size",y="total_bill",data=tips)
plt.title('Line Plot of Total Bill by Size')
plt.show()
```
